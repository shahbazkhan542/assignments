```python
# marksheet program
a = input("name:")
b = input("roll no:")
c = int(input("marks in english "))
d = int(input("marks in math "))
e = int(input("marks in physics "))
f = int(input("marks in chemistry "))
g = int(input("marks in urdu "))
inputtext=""
inputtext += "======marksheet=======\n"
inputtext += "name: "
inputtext += format(a)
inputtext += "\n"
inputtext += "roll no:"
inputtext += format(b)
inputtext += "\n"
inputtext += "subject\t\tmarks obtained\n"
inputtext += "\n"
inputtext += "english\t\t"
inputtext += format(c)
inputtext += "\n"
inputtext += "math\t\t"
inputtext += format(d)
inputtext += "\n"
inputtext += "physics\t\t"
inputtext += format(e)
inputtext += "\n"
inputtext += "chemistry\t"
inputtext += format(f)
inputtext += "\n"
inputtext += "urdu\t\t"
inputtext += format(g)
inputtext += "\n"

tot = c + d + e + f + g
per = (tot / 500) * 100
inputtext += "total obtained "
inputtext += format(tot)
inputtext += "\n"
inputtext += "percentage "
inputtext += format(per)
inputtext += "\n"
if (c < 33 or d < 33 or e < 33 or f < 33 or g < 33):
    inputtext += "fail"

elif (per > 80):
    inputtext += "grade: A+"


elif (per > 70 and per < 80):
    inputtext += "grade: A"

elif (per > 60 and per < 70):
    inputtext += "grade: B"

elif (per > 50 and per < 60):
    inputtext += "grade: C"

elif (per > 40 and per < 50):
    inputtext += "grade: D"

elif (per > 33 and per < 40):
    inputtext += "grade: E"

elif (per > 0 and per < 33):
    inputtext += "grade: F"
f = open("marksheet", "+w")
f.write(inputtext)
f.close()
```
